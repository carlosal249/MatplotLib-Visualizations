```python
# Standard Librays
import pandas as pd
import numpy as np
import datetime 

# Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Reading the data
data = pd.read_csv('monitoramento_tempo.csv', sep=',')
data['data'] = pd.to_datetime(data['data'])
data.head()
```

```python
plt.figure(figsize=(12,8))
plt.plot(data['data'], data['temperatura']);
plt.title('Variaçao da Temperatura X Tempo');
```

```python
# Definindo o tamanho da figura
fig = plt.figure(figsize=(20, 12));

# Criando as figuras
eixo = fig.add_axes([0,0,1,1]);
eixo2 = fig.add_axes([0.7, 0.65, 0.3, 0.3])

# Defindindo o Grafico da Temperatura em Maio
eixo.plot(data['data'], data['temperatura'], color = 'g'); # lw=4 ls='dotted'
eixo.set_xlim(datetime.datetime(2014,5,1),datetime.datetime(2014,6,1))
eixo.set_ylim(270, 320)
eixo.legend(['temperatura'], loc='lower right', fontsize=15)
eixo.set_title('Variaçao da Temperatura em Maio', fontsize=25, pad=20);
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.grid(True)

# Selecionando a Variaçao das datas
azul_esquerda = data['data'] < datetime.datetime(2014,5,1)
azul_direita = data['data'] > datetime.datetime(2014,6,1)

# Definindo o Grafico menor
eixo2.plot(data['data'], data['temperatura'], color='g')
eixo2.plot(data[azul_esquerda]['data'], data[azul_esquerda]['temperatura'], color='b')
eixo2.plot(data[azul_direita]['data'], data[azul_direita]['temperatura'], color='b')

# Legendas, labels e titulos do grafico menor
eixo2.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo2.legend(['Temperatura'], loc='best', fontsize=15);
eixo2.set_ylabel('Temperatura', fontsize=10)
eixo2.set_xlabel('Data', fontsize=10)
eixo2.set_title('Variaçao de Temperatura em 2014', pad=15);
eixo2.grid(True)


# Para salvar visualizaçao o 1 Parametro e o nome do objeto que deseja guardar
# e o segundo parametro bbox_inches='tight'  apenas mantem os eixos dos graficos na figura.png
fig.savefig('temperatura.png',bbox_inches='tight')
```

```python
# Inicializando o objeto
fig = plt.figure(figsize=(15, 8));
eixo = fig.add_axes([0,0,1,1]);

eixo.plot(data['data'], data['temperatura']);
eixo.set_title('Variaçao da Temperatura X Tempo', fontsize=25);
eixo.set_ylabel('Temperatura', fontsize=20);
eixo.set_xlabel('Tempo', fontsize=20);
eixo.legend(['Temperatura'], loc='lower right', fontsize=15);
eixo.grid(True)

# Criando linhas no grafico
# axhline # Linhas horizontais no grafico
# axvline # Linhas verticais no grafico
eixo.axhline(max(data['temperatura']), color='r', linestyle='--');
eixo.axhline(min(data['temperatura']), color='r', linestyle='--');
```

```python
# Inicializando o objeto
fig = plt.figure(figsize=(15, 8));
eixo = fig.add_axes([0,0,1,1]);

eixo.plot(data['data'], data['temperatura']);
eixo.set_title('Variaçao da Temperatura X Tempo', fontsize=25);
eixo.set_ylabel('Temperatura', fontsize=20);
eixo.set_xlabel('Tempo', fontsize=20);
eixo.legend(['Temperatura'], loc='lower right', fontsize=15);
eixo.grid(True)

#Criando pontos de Anotaçoes no grafico
# Selecionando o Ponto onde a temperatura atinge seu valor maximo!
x1 = data['data'][data['temperatura'].idxmax()]
y1 = max(data['temperatura'])

# Anotando o no ponto maximo da temperatura!
eixo.annotate('Temperatura', xy=(x1,y1), fontsize=20)


# Criando linhas no grafico
eixo.axhline(max(data['temperatura']), color='r', linestyle='--');
eixo.axhline(min(data['temperatura']), color='r', linestyle='--');
```

```python
fig = plt.figure(figsize=(10,6))
eixo = fig.add_axes([0,0,1,1])

# Marcando a media no Grafico
eixo.axvline(data['temperatura'].mean(), color='r', linestyle='--');
eixo.annotate('$Média$', xy=(286, 0.05))

eixo.hist(data['temperatura'], bins=30, density=True)
eixo.set_title('Histograma', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da pétala', fontsize=15)
eixo.grid(True)
```

```python
fig = plt.figure(figsize=(10,6))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(data['temperatura'], bins=30)
# O parametro Pad remete ao Padding, fazendo o texto 'subir' um pouco do grafico
eixo.set_title('Histograma da Temperatura', fontsize=15, pad=10)
eixo.set_xlabel('Temperatura em K', fontsize=15)
eixo.grid(True)

# Extraindo valores da media e do desvio Padrao
mu, sigma = data['temperatura'].mean(), data['temperatura'].std()

# colocar entre $.. $ para indicar uma 'formula'
eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
             xy=(300, 2500), fontsize=15);

eixo.axvline(data['temperatura'].mean(), color='r', linestyle='--');
# eixo.axhline(data['pressão'].mean(), color='r', linestyle='--');

fig.savefig('histograma_temp.png', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(10,6))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(data['pressão'], bins=30)
eixo.set_title('Histograma da pressão', fontsize=15, pad=10)
eixo.set_xlabel('pressão', fontsize=15)
eixo.grid(True)

mu, sigma = data['pressão'].mean(), data['pressão'].std()

eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
             xy=(1040, 3000), fontsize=15);

eixo.axvline(data['pressão'].mean(), color='r', linestyle='--');
fig.savefig('histograma_pres.png', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(10,6))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(data['humidade'], bins=30)
eixo.set_title('Histograma da humidade', fontsize=15, pad=10)
eixo.set_xlabel('humidade', fontsize=15)
eixo.grid(True)

mu, sigma = data['humidade'].mean(), data['pressão'].std()

eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
             xy=(60, 3000), fontsize=15);


eixo.axvline(data['humidade'].mean(), color='r', linestyle='--');
fig.savefig('histograma_humid.png', bbox_inches='tight')
```

```python
from PIL import Image

largura, altura = Image.open('histograma_humid.png').size

imagem_combinada = Image.new("RGB", (3*largura, altura))
intervalo = 0

for imagem in map(Image.open, ['histograma_humid.png', 'histograma_pres.png',
                              'histograma_temp.png']):
    imagem_combinada.paste(imagem, (intervalo, 0))
    intervalo += largura

imagem_combinada.save('imagem_combinada.png')
```
